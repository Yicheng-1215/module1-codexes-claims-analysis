```python
#import inpatient data
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Yicheng-1215/module1-codexes-claims-analysis/main/hha.csv', sep='|')
print(df.columns)
```

```python
#identify the medical codex columns
codex_columns = [
    'PRNCPAL_DGNS_CD',
    'ICD_DGNS_CD1', 'ICD_DGNS_CD2', 'ICD_DGNS_CD3', 'ICD_DGNS_CD4', 'ICD_DGNS_CD5', 'ICD_DGNS_CD6', 'ICD_DGNS_CD7', 'ICD_DGNS_CD8', 'ICD_DGNS_CD9', 'ICD_DGNS_CD10',
    'FST_DGNS_E_CD',
    'ICD_DGNS_E_CD1', 'ICD_DGNS_E_CD2', 'ICD_DGNS_E_CD3', 'ICD_DGNS_E_CD4',
    'HCPCS_CD',
    'HCPCS_1ST_MDFR_CD', 'HCPCS_2ND_MDFR_CD'  #
]
```

```python
#Calculate unique value
for column in codex_columns:
    print(f"Frequency of unique values in {column}:")
    print(df[column].value_counts())
```

```python
#Check any missing or null values in the codex-related columns and handle them appropriately
missing_data = df[codex_columns].isnull().sum()
print("Missing data in codex columns:")
print(missing_data)
df[codex_columns] = df[codex_columns].fillna('Unknown')
```

```python
#Find most common code
for column in codex_columns:
    print(f"Most common codes in {column}:")
    print(df[column].value_counts().head(5))
```

```python
#Find most common region
if 'PRVDR_STATE_CD' in df.columns:
    print("Top 3 diagnosis codes by provider state:")
    state_icd_summary = df.groupby('PRVDR_STATE_CD')['PRNCPAL_DGNS_CD'].value_counts().groupby(level=0).head(3)
    print(state_icd_summary)
```
